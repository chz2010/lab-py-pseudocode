### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
BEGIN
  READ N
  SET sum = 0
  FOR i = 1 TO N DO
    SET sum = sum + i
  END FOR
  PRINT sum
END

```python
# Add your code solution here
n = int(input("Enter N: "))

total = 0
for i in range(1, n + 1):
    total += i

print(f"Sum of first {n} natural numbers = {total}")
```

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

<!-- Add your pseudocode here -->
EGIN
  READ string
  SET count = 0
  SET vowels = "aeiouAEIOU"
  FOR each character IN string DO
    IF character IN vowels THEN
      SET count = count + 1
    END IF
  END FOR
  PRINT count
END


```python
# Add your code solution here

string = input("Enter a string: ")
vowels = "aeiouAEIOU"
count  = 0

for char in string:
    if char in vowels:
        count += 1

print(f"Number of vowels: {count}")
```

### Bonus Exercise 6: Reverse a String

Write pseudocode to reverse a given string and print it out.

<!-- Add your pseudocode here -->
BEGIN
  READ string
  SET reversed = ""
  SET index = LENGTH(string) - 1
  WHILE index >= 0 DO
    SET reversed = reversed + string[index]
    SET index = index - 1
  END WHILE
  PRINT reversed
END

```python
# Add your code solution here
string   = input("Enter a string: ")
reversed_string = ""
index    = len(string) - 1

while index >= 0:
    reversed_string += string[index]
    index -= 1

print(f"Reversed: {reversed_string}")
```

### Bonus Exercise 8: Count Words in a String

Write pseudocode that counts the words in a given string and prints the result.

<!-- Add your pseudocode here -->
BEGIN
  READ string
  SET count = 0
  SET in_word = FALSE
  FOR each character IN string DO
    IF character is NOT a space THEN
      IF in_word = FALSE THEN
        SET count = count + 1
        SET in_word = TRUE
      END IF
    ELSE
      SET in_word = FALSE
    END IF
  END FOR
  PRINT count
END

```python
# Add your code solution here
string  = input("Enter a string: ")
count   = 0
in_word = False

for char in string:
    if char != " ":
        if not in_word:
            count  += 1
            in_word = True
    else:
        in_word = False

print(f"Word count: {count}")
```

BEGIN

  ELSE

END

### Bonus Exercise 10: FizzBuzz Problem

Write pseudocode for the FizzBuzz problem where you print numbers from 1 to N but replace multiples of three with "Fizz" and multiples of five with "Buzz".

<!-- Add your pseudocode here -->
BEGIN
  READ N
  FOR i = 1 TO N DO
    IF i MOD 3 = 0 AND i MOD 5 = 0 THEN
      PRINT "FizzBuzz"
    ELSE IF i MOD 3 = 0 THEN
      PRINT "Fizz"
    ELSE IF i MOD 5 = 0 THEN
      PRINT "Buzz"
    ELSE
      PRINT i
    END IF
  END FOR
END

```python
# Add your code solution here
n = int(input("Enter N: "))

for i in range(1, n + 1):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
```
